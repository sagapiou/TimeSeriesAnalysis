```python
import numpy as np
import pandas as pd
import cufflinks as cf
from plotly import __version__
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot 
import plotly.express as px
from datetime import datetime,date

%matplotlib inline
init_notebook_mode(connected=True)
cf.set_config_file(sharing='private',theme='pearl',offline=True)
#print("Cufflinks Version : {}".format(cf.__version__))

df = pd.read_csv('./data/btc-usd-max.csv',index_col=0,parse_dates=True )
df = df.resample('D').first()
df['price'].fillna(method='ffill', inplace=True)
# to get the log price
# df['price'] = np.log(df['price'])
df2 = pd.read_csv('./data/eth-usd-max.csv',index_col=0,parse_dates=True )
df2 = df.resample('D').first()
df2['price'].fillna(method='ffill', inplace=True)
# to get the log price
# df['price'] = np.log(df['price'])
df.head()
df.head()
```

```python
from statsmodels.graphics.tsaplots import month_plot,quarter_plot
```

```python
# quarter plot BTC
dfq = df['price'].resample(rule='Q').mean()
quarter_plot(dfq);
```

```python
# monthly plot BTC
dfq = df['price'].resample(rule='M').mean()
month_plot(dfq);
```

```python
# quarter plot ETH
dfq = df2['price'].resample(rule='Q').mean()
quarter_plot(dfq);
```

```python
# monthly plot ETH
dfm = df2['price'].resample(rule='M').mean()
month_plot(dfm);
```
